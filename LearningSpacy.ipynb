```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

**TOKENIZATION**

```python
doc=nlp("They're linguists!")#list of words/tokens - TOKENIZATION
text="They're linguists!"
print(doc)
print(text)
print(len(doc))
print(len(text))
print(doc[0])
print(text[0])
for token in doc:
  print(token)
  print(token.text)
for token in text:
  print(token)
```

```python
tokens=[token.text for token in doc]
print(tokens)
```

**LEMMATIZATION**
Turning words into their root form, also called a lemma.

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

**PART-OF-SPEECH TAGGING** Assigning a part of speech, such as a noun or verb, to each word in a text.

```python
 poses=[token.pos_ for token in doc]
 poses
```

**NAMED ENTITY RECOGINITION**
Identifying and classifying named entities, such as peope, organizations, and locations, in a text.


```python
doc=nlp("Noam Chomsky was born in the US in 1928.")
```

```python
print(doc.ents)
print(doc.ents[0].text)
print(doc.ents[0].label)
```

```python
for token in doc.ents:
  print(token.)
```
