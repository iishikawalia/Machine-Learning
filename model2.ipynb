```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([[0,0],[0,1],[1,0],[1,1]],dtype=torch.float32)
y = torch.tensor([[0],[1],[1],[1]],dtype=torch.float32)
```

```python
model=nn.Linear(2,1)
loss_fn=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
print(X.shape)
print(y.shape)
```

```python
epochs=1000
for epoch in range(epochs):
  model.train()
  y_pred=model(X)
  loss=loss_fn(y_pred,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  if epoch % 100 == 0:
    print(loss.item())
```

```python
print(model.weight)
print(model.bias)
```
