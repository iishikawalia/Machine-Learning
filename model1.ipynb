```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([[1],[2],[3],[4]],dtype=torch.float32)
y = torch.tensor([[3],[5],[7],[9]],dtype=torch.float32)
```

```python
model=nn.Linear(1,1)
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.01
)
```

```python
epochs=1000
for epoch in range(epochs):
  predictions=model(X)
  loss=loss_fn(predictions,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
print(model.weight)
print(model.bias)
```
