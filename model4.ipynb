```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([[1.0],[2.0],[3.0],[4.0]],dtype=torch.float32)
y = torch.tensor([[1.0],[4.0],[7.0],[10.0]],dtype=torch.float32)
```

```python
model=nn.Linear(1,1)
loss_fn=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs=1000
for epoch in range(epochs):
  model.train()
  y_pred=model(X)
  loss=loss_fn(y_pred,y)
  loss.backward()
  optimizer.step()
  optimizer.zero_grad()
  if epoch%100==0:
    print(f"epoch {epoch}: loss = {loss.item():.4f}")
```

```python
print(model.parameters)
print(model.weight)
print(model.bias)
```

```python
with torch.inference_mode():
    print(model(X))
```
